Subtask 3.4: Display the names of top 5 directors based on IMDb ratings with a minimum of 5 movies (Top Directors Playlist) (4 marks)


```python
# Write your code here
x=movies.groupby("director_name")[['imdb_score','movie_title']].agg({"movie_title":"count","imdb_score":np.mean})
top_5_d=x.loc[x.movie_title>=5].sort_values(by=["imdb_score"],ascending= False).head(5)
top_5_d
```

-  ### Subtask 3.5.1: Find the most popular genre of movies (by highest average rating) (4 marks)

```python
# Write your code here
#code to extract the primary genre of the movie
movie_genres = movies.copy()
movie_g = movie_genres['genres'].str.split('|', expand = True).rename(columns = lambda x: "string"+str(x+1))
movie_genres = pd.concat([movie_genres,movie_g],axis =1)
movie_genres = movie_genres.drop(['string2','string3','string4','string5','string6','string7','string8'], axis=1)
movie_genres.rename(columns={'string1': 'genre_1'}, inplace=True)
movie_genres.head()
```

```python
#code to extract the most popular genre of movies by highest average IMDB rating
genre_popular = movie_genres.pivot_table(index='genre_1',values='imdb_score',aggfunc = 'mean')
genre_popular = pd.DataFrame(genre_popular.to_records())
genre_popular = genre_popular.sort_values(by="imdb_score",ascending=False)
genre_popular.head(1)
```

-  ### Subtask 3.5.2: Find the most popular genre of movies (by total gross) (4 marks)

```python
# Write your code here

genre_popular_gross = movie_genres.pivot_table(index='genre_1',values='gross',aggfunc = 'sum')
genre_popular_gross = pd.DataFrame(genre_popular_gross.to_records())
genre_popular_gross = genre_popular_gross.sort_values(by="gross",ascending=False)
genre_popular_gross.head(1)
```
